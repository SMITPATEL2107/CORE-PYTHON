```python
class car():
    def __init__(self,name):
        self.name=name
    def show(self):
        print(self.name)

class detail(car):
    def __init__(self,name,typpe):
        car.__init__(self,name)
        self.typpe=typpe
    
    def show1(self):
        self.show()
        print(self.typpe)

class sedan(detail):
    def __init__(self,name,typpe,colour):
        detail.__init__(self,name,typpe)
        self.colour=colour
    
    def show2(self):
        self.show1()
        print(self.colour)

class model(car):
    def __init__(self,name,modelname):
        car.__init__(self,name)
        self.modelname=modelname
    
    def show3(self):
        self.show()
        print(self.modelname)
        
class amount(model):
    def __init__(self,name,modelname,price):
        model.__init__(self,name,modelname)
        self.price=price
        
    def show4(self):
        self.show3()
        print(self.price)


name=input("enter name of car:")
typpe=input("enter name of compony:")
color=input("enter type of colour:")
s=sedan(name,typpe,color)
s.show2()
m=input("enter model year:")
p=input("enter price:")
a=amount(name,m,p)
a.show4()
```
